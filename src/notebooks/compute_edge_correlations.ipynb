```python
import numpy as np
import os

nifty_root = '/fs/class-projects/fall2021/cmsc742/c742g002/NiftyGNN/dataset'

for dataset in ['bail', 'credit', 'german']:
    print(os.listdir(os.path.join(nifty_root, dataset)))
    
    if dataset == 'bail':
        is_whites = []
        is_males = []
        ages = []
        is_recs = []
        with open(nifty_root + '/' + dataset + '/' + dataset + '.csv', 'r') as csv_file:
            lines = csv_file.readlines()
            print(len(lines[0].strip().split(',')))
            print(len(lines))
            for line in lines[1:]:
                line = line.strip().split(',')
                is_white = line[0]
                is_male = line[9]
                protected = int(line[13]) / 12
                if protected < 18:
                    age = 0
                elif protected < 25:
                    age = 1
                elif protected < 33:
                    age = 2
                elif protected < 41:
                    age = 3
                elif protected < 61:
                    age = 4
                else:
                    age = 5
                is_rec = line[16]
                is_whites.append(int(is_white))
                is_males.append(int(is_male))
                ages.append(age)
                is_recs.append(int(is_rec))
                
        with open(nifty_root + '/' + dataset + '/' + dataset + '_edges.txt', 'r') as txt_file:
            lines = txt_file.readlines()
            print(len(lines))
            whites_sims = []
            males_sims = []
            ages_sims = []
            recs_sims = []
            for line in lines:
                src_node = line.strip().split(' ')[0]
                src_node = int(10 ** int(src_node.split('+')[1]) * float(src_node.split('e')[0]))
                dst_node = line.strip().split(' ')[1]
                dst_node = int(10 ** int(dst_node.split('+')[1]) * float(dst_node.split('e')[0]))
                
                whites_sims.append(int(is_whites[src_node] == is_whites[dst_node]))
                males_sims.append(int(is_males[src_node] == is_males[dst_node]))
                ages_sims.append(int(ages[src_node] == ages[dst_node]))
                recs_sims.append(int(is_recs[src_node] == is_recs[dst_node]))
        print(np.array(whites_sims).mean())
        print(np.array(males_sims).mean())
        print(np.array(ages_sims).mean())
        print(np.array(recs_sims).mean())
    
    elif dataset == 'credit':
        is_marrieds = []
        is_singles = []
        ages = []
        no_defaults = []
        with open(nifty_root + '/' + dataset + '/' + dataset + '.csv', 'r') as csv_file:
            lines = csv_file.readlines()
            print(len(lines[0].strip().split(',')))
            print(len(lines))
            for line in lines[1:]:
                line = line.strip().split(',')
                is_married = line[1]
                is_single = line[2]
                age = int(line[3])
                no_default = line[0]
                is_marrieds.append(int(is_married))
                is_singles.append(int(is_single))
                ages.append(age)
                no_defaults.append(int(no_default))
                
        with open(nifty_root + '/' + dataset + '/' + dataset + '_edges.txt', 'r') as txt_file:
            lines = txt_file.readlines()
            print(len(lines))
            married_sims = []
            single_sims = []
            ages_sims = []
            no_defaults_sims = []
            for line in lines:
                src_node = line.strip().split(' ')[0]
                src_node = int(10 ** int(src_node.split('+')[1]) * float(src_node.split('e')[0]))
                dst_node = line.strip().split(' ')[1]
                dst_node = int(10 ** int(dst_node.split('+')[1]) * float(dst_node.split('e')[0]))
                
                married_sims.append(int(is_marrieds[src_node] == is_marrieds[dst_node]))
                single_sims.append(int(is_singles[src_node] == is_singles[dst_node]))
                ages_sims.append(int(ages[src_node] == ages[dst_node]))
                no_defaults_sims.append(int(no_defaults[src_node] == no_defaults[dst_node]))
        print(np.array(married_sims).mean())
        print(np.array(single_sims).mean())
        print(np.array(ages_sims).mean())
        print(np.array(no_defaults_sims).mean())
        
    elif dataset == 'german':
        genders = []
        foreigns = []
        singles = []
        ages = []
        no_defaults = []
        with open(nifty_root + '/' + dataset + '/' + dataset + '.csv', 'r') as csv_file:
            lines = csv_file.readlines()
            print(len(lines))
            print(len(lines[0].strip().split(',')))
            for line in lines[1:]:
                line = line.strip().split(',')
                gender = line[1]
                if gender == 'Male':
                    gender = 1
                else:
                    gender = 0
                foreign = line[2]
                single = line[3]
                protected = int(line[4])
                if protected < 18:
                    age = 0
                elif protected < 25:
                    age = 1
                elif protected < 33:
                    age = 2
                elif protected < 41:
                    age = 3
                elif protected < 61:
                    age = 4
                else:
                    age = 5
                no_default = line[0]
                if no_default == '-1':
                    no_default = 0
                genders.append(int(gender))
                foreigns.append(int(foreign))
                singles.append(int(single))
                ages.append(age)
                no_defaults.append(int(no_default))
                
        with open(nifty_root + '/' + dataset + '/' + dataset + '_edges.txt', 'r') as txt_file:
            lines = txt_file.readlines()
            print(len(lines))
            gender_sims = []
            foreign_sims = []
            single_sims = []
            ages_sims = []
            no_defaults_sims = []
            for line in lines:
                src_node = line.strip().split(' ')[0]
                src_node = int(10 ** int(src_node.split('+')[1]) * float(src_node.split('e')[0]))
                dst_node = line.strip().split(' ')[1]
                dst_node = int(10 ** int(dst_node.split('+')[1]) * float(dst_node.split('e')[0]))
                
                gender_sims.append(int(genders[src_node] == genders[dst_node]))
                foreign_sims.append(int(foreigns[src_node] == foreigns[dst_node]))
                single_sims.append(int(singles[src_node] == singles[dst_node]))
                ages_sims.append(int(ages[src_node] == ages[dst_node]))
                no_defaults_sims.append(int(no_defaults[src_node] == no_defaults[dst_node]))
        print(np.array(gender_sims).mean())
        print(np.array(foreign_sims).mean())
        print(np.array(single_sims).mean())
        print(np.array(ages_sims).mean())
        print(np.array(no_defaults_sims).mean())
        
```

```python
import numpy as np
import os

nba_dir = '../../dataset/NBA/'

print(os.listdir(nba_dir))

lines = []
with open(nba_dir + 'nba.csv', 'r') as csv_file:
    lines = csv_file.readlines()

user_idx = 0
salary_idx = 1
age_idx = 2
country_idx =37

salaries = {}
ages = {}
countries = {}
for line in lines[1:]:
    user_id = int(line.split(',')[user_idx])
    salary = line.split(',')[salary_idx]
    protected = int(float(line.split(',')[age_idx]))
    country = line.split(',')[country_idx]
    
    if salary == '-1':
        salary = 0
    
    
    if protected < 22:
        age = 0
    elif protected < 26:
        age = 1
    elif protected < 30:
        age = 2
    elif protected < 33:
        age = 3
    else:
        age = 4
        
    salaries[user_id] = int(salary)
    ages[user_id] = age
    countries[user_id] = int(country)

lines = []
salary_sims = []
age_sims = []
country_sims = []
with open(nba_dir + 'nba_relationship.txt', 'r') as txt_file:
    lines = txt_file.readlines()
    for line in lines:
        line = line.strip().split('\t')
        src_node = int(line[0])
        dst_node = int(line[1])
        
        salary_sims.append(int(salaries[src_node] == salaries[dst_node]))
        age_sims.append(int(ages[src_node] == ages[dst_node]))
        country_sims.append(int(countries[src_node] == countries[dst_node]))
print('salary', np.array(salary_sims).mean())
print('age', np.array(age_sims).mean())
print('country', np.array(country_sims).mean())
```
